```python
from scipy import misc
import numpy as np
import imageio
import cv2

def periodic_noise_level_1(eachName):
    %cd '/content/drive/MyDrive/training-jaggu/dcscn-super-resolution/data/train60p_R_4'# Training Dataset Folder path
    orig = imageio.imread(eachName)
    sh = orig.shape[0], orig.shape[1], orig.shape[2]
    noise = np.zeros(sh, dtype='float64')

    X, Y, Z = np.meshgrid(range(0, sh[0]), range(0, sh[1]),  range(0, sh[2]))

    A = 50
    u0 = 45
    v0 = 50
    p0 = 40

    noise += A*np.sin(X*u0 + Y*v0 + Z*p0)
    A = -36
    u0 = -45
    v0 = -50
    p0 =  -40

    noise += A*np.sin(X*u0 + Y*v0 + Z*p0)

    noiseada = orig+noise
    noiseada = cv2.normalize(noiseada, None, 0, 255, cv2.NORM_MINMAX, cv2.CV_8U)
    %cd '/content/drive/MyDrive/Saujas/deblur-gan/Dataset-Ayush/train/A'# A folder path
    imageio.imsave(eachName, noiseada)

def periodic_noise_level_2(eachName):
    %cd '/content/drive/MyDrive/training-jaggu/dcscn-super-resolution/data/train60p_R_4'# Training Dataset Folder path
    orig = imageio.imread(eachName)
    sh = orig.shape[0], orig.shape[1], orig.shape[2]
    noise = np.zeros(sh, dtype='float64')

    X, Y, Z = np.meshgrid(range(0, sh[0]), range(0, sh[1]),  range(0, sh[2]))

    A = 50
    u0 = 45
    v0 = 50
    p0 = 40

    noise += A*np.cos(X*u0 + Y*v0 + Z*p0)
    A = -36
    u0 = -45
    v0 = -50
    p0 =  -40

    noise += A*np.cos(X*u0 + Y*v0 + Z*p0)

    noiseada = orig+noise
    noiseada = cv2.normalize(noiseada, None, 0, 255, cv2.NORM_MINMAX, cv2.CV_8U)
    %cd '/content/drive/MyDrive/Saujas/deblur-gan/Dataset-Ayush/train/A'# A folder path
    imageio.imsave(eachName, noiseada)

def periodic_noise_level_3(eachName):
    %cd '/content/drive/MyDrive/training-jaggu/dcscn-super-resolution/data/train60p_R_4'# Training Dataset Folder path
    orig = imageio.imread(eachName)
    sh = orig.shape[0], orig.shape[1], orig.shape[2]
    noise = np.zeros(sh, dtype='float64')

    X, Y, Z = np.meshgrid(range(0, sh[0]), range(0, sh[1]),  range(0, sh[2]))

    A = 50
    u0 = 45
    v0 = 50
    p0 = 40

    noise += A*np.sin(X*u0 + Y*v0 + Z*p0)
    A = -36
    u0 = -45
    v0 = -50
    p0 =  -40

    noise += A*np.cos(X*u0 + Y*v0 + Z*p0)

    noiseada = orig+noise
    noiseada = cv2.normalize(noiseada, None, 0, 255, cv2.NORM_MINMAX, cv2.CV_8U)
    %cd '/content/drive/MyDrive/Saujas/deblur-gan/Dataset-Ayush/train/A'# A folder path
    imageio.imsave(eachName, noiseada)

def periodic_noise_level_4(eachName):
    %cd '/content/drive/MyDrive/training-jaggu/dcscn-super-resolution/data/train60p_R_4' # Training Dataset Folder path
    orig = imageio.imread(eachName)
    sh = orig.shape[0], orig.shape[1], orig.shape[2]
    noise = np.zeros(sh, dtype='float64')

    X, Y, Z = np.meshgrid(range(0, sh[0]), range(0, sh[1]),  range(0, sh[2]))

    A = 50
    u0 = 45
    v0 = 50
    p0 = 40

    noise += A*np.cos(X*u0 + Y*v0 + Z*p0)
    A = -36
    u0 = -45
    v0 = -50
    p0 =  -40

    noise += A*np.sin(X*u0 + Y*v0 + Z*p0)

    noiseada = orig+noise
    noiseada = cv2.normalize(noiseada, None, 0, 255, cv2.NORM_MINMAX, cv2.CV_8U)
    %cd '/content/drive/MyDrive/Saujas/deblur-gan/Dataset-Ayush/train/A'# A folder path
    imageio.imsave(eachName, noiseada)    
```
